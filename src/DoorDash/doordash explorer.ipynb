````python
def timedelta_str(str_a, str_b, formatter='%Y-%m-%d %H:%M:%S'):
    datetime1 = datetime.strptime(str_a, formatter)
    datetime2 = datetime.strptime(str_b, formatter)
    timedelta = datetime2 - datetime1
    return timedelta.seconds
````

````python
datetime1 = datetime.strptime(data_df.iloc[0]['created_at'], '%Y-%m-%d %H:%M:%S')
datetime2 = datetime.strptime(data_df.iloc[0]['actual_delivery_time'], '%Y-%m-%d %H:%M:%S')
timedelta = datetime2 - datetime1
````

### some insights
* we can drop column `actual_delivery_time`, `estimated_store_to_consumer_driving_duration`, remove should be fine
* columns `total_onshift_dashers`, `total_busy_dashers`, `total_outstanding_orders` have same ratio of nan, maybe they are nan at same rows. And the ratio is quite high, maybe not so good to just remove
* for `market_id`, `order_protocol`, `store_primary_category` should be fine to just consider as nan and one-hot encoding

following code is use to prove

So for preprocessing, we do following simple approach
* remove rows with nan in columns `actual_delivery_time`, `estimated_store_to_consumer_driving_duration`
* consider categorical nan as a new category in `market_id`, `order_protocol`, `store_primary_category` and calculate one-hot coding
* in columns `total_onshift_dashers`, `total_busy_dashers`, `total_outstanding_orders`, it will depends on the model we use. we might also derive some new features from this, such as percentage  

````python
# remove those with on actual_delivery_time
new_df = data_df.dropna(subset=['actual_delivery_time', 'estimated_store_to_consumer_driving_duration'])
new_df
````

````python
new_df['total_seconds'] = new_df.apply(lambda x: timedelta_str(x['created_at'], x['actual_delivery_time']), axis=1)
````

````python
new_df
````

````python
# check unique value for categorical columns
category_columns = ["market_id", "store_id", 'store_primary_category', 'order_protocol']
unique_value = {}
for col in category_columns:
    unique_value[col] = new_df[col].unique()
unique_value
````

````python
unique_num = {}
for col in unique_value:
    unique_num[col] = len(unique_value[col])
unique_num
````

````python
for col in category_columns:
    new_df[col].fillna('NaN', inplace=True)
````

````python
one_hot_df = new_df.copy()
for col in category_columns:
    one_hot_df = pd.concat([one_hot_df, pd.get_dummies(new_df[col], prefix=col)], axis=1)
one_hot_df.drop(columns=category_columns)
````

````python
final_raw_data = one_hot_df.drop(columns=category_columns)
final_raw_data.shape
````
